```python
import pandas as pd

# 1. Load the files you uploaded
customers = pd.read_csv('olist_customers_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
order_items = pd.read_csv('olist_order_items_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')
translations = pd.read_csv('product_category_name_translation.csv')
```

```python
# 2. Merge them step-by-step
# Combine Orders and Customers to get location data
df = pd.merge(orders, customers, on='customer_id')

# Combine with Order Items to get prices and freight
df = pd.merge(df, order_items, on='order_id')

# Combine with Products to get category IDs
df = pd.merge(df, products, on='product_id')

# Combine with Translations to get English category names
df = pd.merge(df, translations, on='product_category_name')
```

```python
# 3. Clean Date Columns
date_cols = ['order_purchase_timestamp', 'order_delivered_customer_date', 'order_estimated_delivery_date']
for col in date_cols:
    df[col] = pd.to_datetime(df[col])
```

```python
# 4. Feature Engineering (This looks great on a resume!)
# Calculate actual delivery time vs estimated
df['actual_delivery_days'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days
df['revenue'] = df['price'] + df['freight_value']
```

```python
# 5. Export the Final Master File
df.to_csv('olist_master_cleaned.csv', index=False)

print("Success! Your master file is ready.")
```
